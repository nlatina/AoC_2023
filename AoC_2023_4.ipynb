```python
'''day 4'''
import numpy as np
```

```python
with open('data/day4.txt', 'r') as day4:
    data = [d.strip() for d in day4]
data = [line.split(':')[1].split('|') for line in data]
```

```python
print(data[0])
```

```python
def process(row):
    winning_nums = row[0].split(" ")
    while "" in winning_nums:
        winning_nums.remove("")
        
    inventory = row[1].split(" ")
    while "" in inventory:
        inventory.remove("")
    return(set(winning_nums), set(inventory))
```

```python
#print(set(data[0][0]) - set(' '))
numbers = []
for i in range(len(data)):
    winning, inventory = process(data[i])
    numbers.append(len(winning & inventory))
    
print(numbers)
points = [2**(n-1) for n in numbers if n > 0]
print(sum(points))
```

```python
temp = np.ones(len(numbers))

extended_numbers = [list(pair) for pair in zip(numbers, temp)]

extended_numbers.append([0, 0])

for i in range(len(numbers)):
    for n in range(i+1, i+1+numbers[i]):
        extended_numbers[n][1] += extended_numbers[i][1]

print(extended_numbers)
total_copies = [num for _, num in extended_numbers]
print(sum(total_copies))
```
